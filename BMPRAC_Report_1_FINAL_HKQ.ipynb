# Background & Rationale

Cardiovascular disease (CVD) is the leading cause of death globally, accounting for nearly 18 million deaths annually, around 32% of global deaths (WHO, 2021). According to the American Heart Association, almost half of all U.S. adults have some form of cardiovascular disease, and rates continue to rise due to the growing prevalence of modifiable risk factors such as obesity, hypertension, and unhealthy lipid profiles (Benjamin et al., 2019). Among the many risk indicators, systolic blood pressure (BPXSY), high-density lipoprotein (HDL), and low-density lipoprotein (LDL) cholesterol are widely accepted as key biomarkers of cardiovascular health.

Elevated systolic BP and LDL, with low levels of HDL, are strongly associated with increased risk of stroke and heart attacks. Both public health interventions and treatment guidelines regularly target these biomarkers, which are used to assess the likelihood of future cardiovascular events. Over the years, the rise in obesity rates across all age groups has contributed significantly to the burden of CVD. That is why it is essential to understand how body mass index (BMI) affects the general population’s cardiovascular health, with it becoming more urgent than ever to understand the link between BMI and cardiovascular biomarkers, due to the US adult population’s obesity rate exceeding 40% (CDC, 2023).

Since CVD develops gradually (Després, 2012), identifying these changes in midlife can guide public health efforts focused on healthy weight, lifestyle change, and targeted screening.


Biologically, elevated BMI contributes to hypertension and atherosclerosis through mechanisms like increased vascular resistance, altered lipid metabolism, and chronic inflammation (Després, 2012; Ference et al., 2017). This study examines how BMI relates to systolic BP, HDL, and LDL, while accounting for confounders—supporting precision prevention and better CVD risk prediction models.

```python
import pandas as pd
import os
from glob import glob
# from scipy.io import saveload # Removed as it's not used and causes ImportError

# Define the years and corresponding file patterns for the main NHANES data
nhanes_years = {
    2007: (2007, 2008),
    2009: (2009, 2010),
    2011: (2011, 2012),
    2013: (2013, 2014),
    2015: (2015, 2016),
    2017: (2017, 2018)
}

# Define the file patterns for each data component
file_patterns = {
    'DEMO': '/content/NHANES/DEMO_{start_year}-{end_year}.xpt', # Updated file extension
    'BMI': '/content/NHANES/BMI_{start_year}-{end_year}.xpt',
    'HDL': '/content/NHANES/HDL_{start_year}-{end_year}.xpt',
    'LDL': '/content/NHANES/LDL_{start_year}-{end_year}.xpt',
    'PA': '/content/NHANES/PA_{start_year}-{end_year}.xpt',
    'SBP': '/content/NHANES/SBP_{start_year}-{end_year}.xpt',
    'Sodium_Day1': '/content/NHANES/Sodium_{start_year}-{end_year}(Day1).xpt',
    'Sodium_Day2': '/content/NHANES/Sodium_{start_year}-{end_year}(Day2).xpt',
    'Hypertension': '/content/NHANES/Hyp_{start_year}-{end_year}.xpt'
}


# Load and merge data from different cycles and components
all_data = pd.DataFrame()

for year, (start_year, end_year) in nhanes_years.items():
    year_data = pd.DataFrame()
    for component, pattern in file_patterns.items():
        file_path = pattern.format(start_year=start_year, end_year=end_year)
        try:
            df = pd.read_sas(file_path)
            if year_data.empty:
                year_data = df
            else:
                # Merge on SEQN, assuming it's the common identifier
                year_data = pd.merge(year_data, df, on='SEQN', how='outer')
            print(f"Loaded and merged {component} data from {file_path}")
        except FileNotFoundError:
            print(f"File not found for {component} data for {start_year}-{end_year}: {file_path}")
        except Exception as e:
            print(f"Error loading file {file_path}: {e}")

    if not year_data.empty:
        all_data = pd.concat([all_data, year_data], ignore_index=True)
        print(f"Concatenated data for {start_year}-{end_year}")


# Check the columns in the merged data before proceeding
print("\nColumns in the merged data:")
print(all_data.columns)


# Now proceed with filtering and renaming columns from all_data
columns_to_keep_and_rename = {
    'SEQN': 'SEQN',
    'RIDAGEYR': 'Age',
    'RIAGENDR': 'Gender',
    'RIDRETH1': 'Ethnicity',
    'INDHHIN2': 'Annual Household Income',
    'DR1TSODI': 'Sodium Intake/mg (Day1)', # Assuming these are the correct column names for sodium
    'DR2TSODI': 'Sodium Intake/mg (Day2)',
    'BPXSY1': 'Systolic Blood Pressure/mmHg', # Assuming this is the correct column name for SBP
    'BMXBMI': 'Body Mass Index (BMI)', # Assuming this is the correct column name for BMI
    'LBDLDL': 'LDL-cholesterol (mg/dL)', # Assuming this is the correct column name for LDL
    'LBDHDD': 'HDL-Cholesterol (mg/dL)', # Assuming this is the correct column name for HDL
    'PAQ665': 'Vigorous-intensity activities', # Assuming this is the correct column name for PA
    'BPQ040A': 'Taking hypertension medication' # Assuming this is the correct column name for hypertension medication
}

# Identify existing columns in the merged dataframe that match the keys in columns_to_keep_and_rename
existing_cols = {col: [c for c in all_data.columns if c.startswith(col) and (c == col or (len(c) > len(col) and c[len(col):].isdigit()))]
                 for col in columns_to_keep_and_rename.keys()}

# Prioritize columns without suffixes if they exist, otherwise take the first available
cols_to_select = {}
for col_key, potential_cols in existing_cols.items():
    if col_key in all_data.columns:
        cols_to_select[col_key] = col_key
    elif potential_cols:
        cols_to_select[col_key] = potential_cols[0]
    # If neither exists, the column will not be selected, which will be handled below


# Create a dictionary for renaming based on the selected columns
rename_dict = {selected_col: columns_to_keep_and_rename[original_col_key]
               for original_col_key, selected_col in cols_to_select.items()}

# Filter the DataFrame to keep only the selected columns and rename them
# Ensure all selected columns actually exist in the DataFrame before proceeding
cols_that_exist_and_selected = [col for col in cols_to_select.values() if col in all_data.columns]
filtered_merged_data = all_data[cols_that_exist_and_selected].copy()

# Rename the columns using the prepared rename_dict, but only for columns that exist in the filtered_merged_data
filtered_merged_data = filtered_merged_data.rename(columns={k: v for k, v in rename_dict.items() if k in filtered_merged_data.columns})


print("\nFiltered and Renamed Data:")
display(filtered_merged_data.head())
print("\nShape of the filtered and renamed data:", filtered_merged_data.shape)


# Now perform the categorical variable recoding on the filtered_merged_data
final_merged_data = filtered_merged_data.copy() # Update final_merged_data to continue processing

# Recode categorical variables only if the columns exist
if 'Gender' in final_merged_data.columns:
    final_merged_data['Gender'] = final_merged_data['Gender'].replace({1.0: 'Male', 2.0: 'Female'})
if 'Ethnicity' in final_merged_data.columns:
    final_merged_data['Ethnicity'] = final_merged_data['Ethnicity'].replace({
        1.0: 'Mexican American',
        2.0: 'Other Hispanic',
        3.0: 'Non-Hispanic White',
        4.0: 'Non-Hispanic Black',
        5.0: 'Other Race - Including Multi-Racial'
    })
if 'Annual Household Income' in final_merged_data.columns:
    final_merged_data['Annual Household Income'] = final_merged_data['Annual Household Income'].replace({
        1.0: '$ 0 to $ 4,999',
        2.0: '$ 5,000 to $ 9,999',
        3.0: '$10,000 to $14,999',
        4.0: '$15,000 to $19,999',
        5.0: '$20,000 to $24,999',
        6.0: '$25,000 to $34,999',
        7.0: '$35,000 to $44,999',
        8.0: '$45,000 to $54,999',
        9.0: '$55,000 to $64,999',
        10.0: '$65,000 to $74,999',
        12.0: 'Over $20,000',
        13.0: 'Under $20,000',
        14.0: '$75,000 to $99,999',
        15.0: '$100,000 and Over',
        77.0: 'Refused',
        99.0: "Don't know"
    })
if 'Vigorous-intensity activities' in final_merged_data.columns:
    final_merged_data['Vigorous-intensity activities'] = final_merged_data['Vigorous-intensity activities'].replace({
        1.0: 'Yes',
        2.0: 'No',
        7.0: 'Refused',
        9.0: "Don't Know"
    })
if 'Taking hypertension medication' in final_merged_data.columns:
    final_merged_data['Taking hypertension medication'] = final_merged_data['Taking hypertension medication'].replace({
        1.0: 'Yes',
        2.0: 'No',
        7.0: 'Refused',
        9.0: "Don't Know"
    })


print("\nDataFrame with recoded categorical variables:")
display(final_merged_data.head())


# Drop rows with NaN values
final_merged_data_cleaned = final_merged_data.dropna().copy()

print("\nDataFrame after removing rows with NaN values:")
display(final_merged_data_cleaned.head())
print("\nShape of the cleaned DataFrame:", final_merged_data_cleaned.shape)

# Check if sodium intake columns exist before calculating the average
sodium_cols = ['Sodium Intake/mg (Day1)', 'Sodium Intake/mg (Day2)']
if all(col in final_merged_data_cleaned.columns for col in sodium_cols):
    final_merged_data_cleaned['Average Sodium Intake (mg)'] = final_merged_data_cleaned[sodium_cols].mean(axis=1)

    # Drop the original day 1 and day 2 sodium columns
    final_merged_data_cleaned = final_merged_data_cleaned.drop(columns=sodium_cols)
    print("\nDataFrame with 'Average Sodium Intake (mg)' column:")
    display(final_merged_data_cleaned.head())
else:
    print("\nSodium intake columns ('Sodium Intake/mg (Day1)', 'Sodium Intake/mg (Day2)') not found in the cleaned data.")


# Filter for age range 30-60
if 'Age' in final_merged_data_cleaned.columns:
    final_merged_data_cleaned = final_merged_data_cleaned[(final_merged_data_cleaned['Age'] >= 30) & (final_merged_data_cleaned['Age'] <= 60)].copy()
    print("\nDataFrame filtered to include ages 30-60:")
    display(final_merged_data_cleaned.head())
    print("\nShape of the filtered DataFrame:", final_merged_data_cleaned.shape)
else:
    print("\n'Age' column not found in the cleaned data. Cannot filter by age.")
```

```python
#Shapiro-wilk test

analysis_columns = [
    'Body Mass Index (BMI)',
    'Average Sodium Intake (mg)',
    'Systolic Blood Pressure/mmHg',
    'HDL-Cholesterol (mg/dL)',
    'LDL-cholesterol (mg/dL)'
]

analysis_data = final_merged_data_cleaned[analysis_columns].copy()

print("Analysis Data:")
display(analysis_data.head())
print("\nShape of the analysis data:", analysis_data.shape)




from scipy.stats import shapiro

variables_for_shapiro = [
    'Body Mass Index (BMI)',
    'Average Sodium Intake (mg)',
    'Systolic Blood Pressure/mmHg',
    'HDL-Cholesterol (mg/dL)',
    'LDL-cholesterol (mg/dL)'
]

print("Shapiro-Wilk Test for Normality:")
for variable in variables_for_shapiro:
    if variable in analysis_data.columns:
        stat, p_value = shapiro(analysis_data[variable])
        print(f"Variable: {variable}")
        print(f"  Test Statistic: {stat:.4f}")
        print(f"  P-value: {p_value:.4f}")
        if p_value < 0.05:
            print("  Interpretation: The distribution is likely not normal (reject the null hypothesis of normality).")
        else:
            print("  Interpretation: The distribution is likely normal (fail to reject the null hypothesis of normality).")
    else:
        print(f"Variable: {variable}")
        print("  Column not found in the DataFrame.")
```
